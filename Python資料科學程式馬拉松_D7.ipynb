# **1.請比較對一個 100 x 100 * 100 的陣列，使用不同方法對每一個元素 +1 的時間比較。**


```python
import numpy as np
import time

a=np.random.randint(0,10,1000000).reshape(100,100,100)
b=a
print(a)
print(b)

print("\n")

###########################################
start=time.process_time() #測時間
#法一:廣播+1
print("法一:廣播+1\n",a+1) 
end=time.process_time() #測時間
print("執行時間=",end-start) #測時間
###########################################

print("\n")

###########################################
start=time.process_time() #測時間
#法二:for loop  +1
with np.nditer(b,op_flags=['readwrite']) as iter:   
   for x in iter:
       x[...]=x+1   
print("法二:for loop  +1\n",b)    
end=time.process_time() #測時間
print("執行時間=",end-start)#測時間
###########################################
```

# **2.如何從一個陣列中，找出出現頻率最高的數值與位置？。**

```python
import numpy as np

a=np.random.randint(0,10,100) #亂數產生100個0~9的數字
count=np.bincount(a)      #統計0~9共發生幾次    
print(a)
print(count)
print("\n")
print("出現頻率最高之數值=",count.argmax())
print("出現頻率=",count[count.argmax()])
print("出現頻率最高之數值位置=",np.where(a==count.argmax()))
```

# **3.如何利用 list(...) 實現 a.tolist() 的效果？試著用程式實作。**

```python
import numpy as np
#a=np.arange(30).reshape(5,3,2)
#a = np.random.randint(10, size=6) 

print("==========================")
a = np.random.randint(10, size=6) 
print("法一:tolist():", a.tolist()) 
print("法二:list(a):", list(a))    
print("\n") 

print("==========================")
b = np.random.randint(10, size=(3,4))
print("法一:tolist():", b.tolist())  
print("法二:list(b):", list(b))    
print("\n") 

print("==========================")
c = np.random.randint(10, size=(2,3,2))
print("法一:tolist():", c.tolist()) 
print("法二:list(c):", list(c))    
print("\n") 

def list_append(test_array):
  if(test_array)!=np.ndarray:
    return test_array
  else:
    new_array=[]
    for obj in test_array:
      new_array.append(tolist(test_array))
    return list(new_array)


print("==========================")
print("法一:tolist():", a.tolist()) 
print("法二:list(a):", list(a))    
print("法三:list_append展開--->list_append(a):", list_append(a)) 
print("\n") 

print("==========================")
print("法一:tolist():", b.tolist()) 
print("法二:list(b):", list(b))    
print("法三:list_append展開--->list_append(b):", list_append(b)) 
print("\n") 

print("==========================")
print("法一:tolist():", c.tolist()) 
print("法二:list(c):", list(c))    
print("法三:list_append展開--->list_append(c):", list_append(c)) 
print("\n") 
```
