# **1.[簡答題] 請問下列四種不同的 DataFrame 選取結果有什麼差異？**
圖片1.jpg

```python
import pandas as pd

df=pd.DataFrame(
        {
    'A':[35,21,18,12,32,13],
    'B':[14,34,23,15,22,25],
    'C':[24,19,12,27,40,17],
    'D':[12,30,26,23,43,10],
    'E':[10,29,23,18,45,14],
},

index=['2013-01-01','2013-01-02','2013-01-03','2013-01-04','2013-01-05','2013-01-06']  #index需在字典外面
)

print(df)
print("\n")

print(df.loc['2013-01-01','A']) #取row=2013-01-01與col=A交集之元素
print("\n")
print(df.loc['2013-01-01',['A','B']]) #取row=2013-01-01與col=A,col=B交集之元素
print("\n")
print(df.loc['2013-01-01':'2013-01-02','A']) #取row=2013-01-01~2013-01-02與col=A交集之元素
print("\n")
print(df.loc['2013-01-01':'2013-01-05','A':'C']) #取row=2013-01-01~2013-01-05與col=A~C交集之元素
print("\n")
```

# **2.請根據提供的資料，選擇出下列的要求：**
圖片2.jpg

```python
print("===============================")
print(df)
print("===============================\n")
print(df.iloc[0:3,:])
print("\n")
print(df.iloc[::2,:])
print("\n")
print(df.iloc[:,-2:])
print("\n")
print(df.iloc[:,::2])
print("\n")
```

# **3.請根據提供的資料，選擇出下列的要求：**
圖片3.jpg

```python
print("===============================")
print(df)
print("===============================\n")
print(df.loc[df.A>20,['A','B','C','D','E']]) #df.loc[[]] 可帶條件,可帶col
print("\n")
print(df.loc[(df.A+df.B>50)])    #df.A....要加()
print("\n")
print(df.loc[(df.A<30)|(df.B>30)])  #df.A....要加()
print("\n")
print(df.loc[df.sum(axis=1)>100])
print("\n")
```
